```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame from the table
df = pd.DataFrame({
    'Time': ['0:30', '1:00', '1:30', '2:00', '2:30', '3:00', '3:30', '4:00', '4:30', '5:00'],
    'Temp_Out': [23, 23.1, 22.9, 22.9, 22.6, 22.7, 22.6, 22.5, 22.2, 21.2],
    'Humidity_out': [34, 33, 33, 33, 34, 32, 32, 32, 37, 37],
    'Dew_Pt.': [6.3, 5.9, 5.8, 5.8, 5.9, 5.1, 5.1, 5, 5.6, 5.9],
    'Wind_Speed': [2.2, 2.7, 2.7, 3.1, 2.7, 2.7, 2.2, 3.1, 4.9, 6.3],
    'Heat_Index': [22.1, 22.1, 21.8, 21.9, 21.4, 21.4, 21.3, 21.2, 19.6, 19.3]
})

# Create a figure and axes
fig, ax = plt.subplots()

# Add the table to the axes
ax.table(cellText=df.values, colLabels=df.columns, loc='center')

# Set the title of the figure
plt.title('Weather Data')

# Set the x-axis label
plt.xlabel('Temp_Out')

# Set the y-axis label
plt.ylabel('Time')

# Show the figure
plt.show()
```
